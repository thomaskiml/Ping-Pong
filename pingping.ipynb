```python
# mass of ping pong ball [g]
m = 0.0027

g = 9.8
pi =  3.1415926535
# radius of pingpong ball [m]
r = 0.019

# height from which the ball si dropped [m]
h = 1

# dag coeficient Cd = D / (q * A) https://scienceworld.wolfram.com/physics/DragCoefficient.html
Cd= 0.5

# ping-pong ball has an average density of 0.0840 g/cm3 so I converted it to SI units kg/m3
rw = 84 

# velocity 

V = 0

# A = will be the eference idea of where the ball should be causing friction
# with the air molecules around it. I will be counting only one half af if it were a flow.

A = pi*2*r**2
```

```python
# this is drag function calculation

def drag(vel):
    F_aero = Cd * A * 0.5 * r * vel^2
    return F_aero

#  buoyancy force F_buoy 
vole = 4/3*pi*r**3 # sphere volume 

F_buoy = vole*g*rw

# gravitational force 
F_g = m*g
```
